```python
from openai import OpenAI
from scraper import fetch_website_contents
from IPython.display import Markdown, display

OLLAMA_BASE_URL = "http://localhost:11434/v1"  # this is like openai's /v1/chat endpoint

ollama = OpenAI(
    base_url=OLLAMA_BASE_URL, api_key="ollama"
)  # api_key is a mandatory field so we just pass in 'ollama'
```

```python
system_prompt = """
You are a helpful FAQ generator. When given the text content of a webpage, you generate exactly 5 clear, likely questions a user might ask about that page — along with short, accurate answers based only on the provided content. Do not use any outside knowledge. If the content does not support an answer, say "This information is not available on the page."
Return the answer in pure markdown format.
"""


def generate_user_prompt(website):
    return f"""
Generate 5 frequently asked questions (FAQs) a user might have after reading this page, along with a short answer for each. Base your answers only on the content above.

Here is the content from a webpage: ${fetch_website_contents(website)}
"""


response = ollama.chat.completions.create(
    model="llama3.2:1b",
    messages=[
        {"role": "system", "content": system_prompt},
        {
            "role": "user",
            "content": generate_user_prompt("https://en.wikipedia.org/wiki/Anthropic"),
        },
    ],
)  # can use "llama3.2:1b" if "llama3.2" is too large.

faqs_generated = response.choices[0].message.content

display(Markdown(faqs_generated))
```
