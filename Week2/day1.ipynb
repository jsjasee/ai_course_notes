## Training vs Inference time scaling

inference time scaling is also known as run time scaling


````python
easy_puzzle = [
    {
        "role": "user",
        "content": "You toss 2 coins. One of them is heads. What's the probability the other is tails? Answer with the probability only.",
    },
]
````

````python
# tiny model, little reasoning -> wrong ans
response = openai.chat.completions.create(
    model="gpt-5-nano", messages=easy_puzzle, reasoning_effort="minimal"
)
# new parameter, 'reasoning_effort'
# previously we used 'stream' and 'response_format'
display(Markdown(response.choices[0].message.content))
````

````python
# tiny model but dial up the reasoning -> correct ans
response = openai.chat.completions.create(
    model="gpt-5-nano", messages=easy_puzzle, reasoning_effort="low"
)
display(Markdown(response.choices[0].message.content))
````

````python
# use a slightly bigger model, but reasoning effort is minimal -> correct ans
response = openai.chat.completions.create(
    model="gpt-5-mini", messages=easy_puzzle, reasoning_effort="minimal"
)
display(Markdown(response.choices[0].message.content))
````

#### Answer explanation

The answer is not half, here's the explanation:

Initially I listed out the 4 possible outcomes - HH, HT, TH, TT.

But your 1/2 count includes TT, but TT is impossible because you’re told one of the coins is heads.

So the remaining possible outcomes are:

- HH
- HT
- TH

Out of these 3 equally likely cases, the “other coin is tails” happens in 2 of them (HT, TH).

So the probability is 2/3.

The only way it would be 1/2 is if a specific coin was identified as heads, like: “The first coin is heads.” Then the possibilities would be only HH or HT, giving 1/2.

## Prompt Caching with OpenAI

For OpenAI:

https://platform.openai.com/docs/guides/prompt-caching

> Cache hits are only possible for exact prefix matches within a prompt. To realize caching benefits, place static content like instructions and examples at the beginning of your prompt, and put variable content, such as user-specific information, at the end. This also applies to images and tools, which must be identical between requests.

- So for chatGPT, put the info that changes regularly (eg. today's date) at the END of the prompt, if you put it at the start of the prompt then it will look very different and caching will not happen.
- for the hamlet example, DON'T put the questions THEN hamlet text, put the questions FIRST then hamlet's text

Cached input is 4X cheaper

https://openai.com/api/pricing/

## And now for some fun - an adversarial conversation between Chatbots..

You're already familar with prompts being organized into lists like:

```
[
    {"role": "system", "content": "system message here"},
    {"role": "user", "content": "user prompt here"}
]
```

In fact this structure can be used to reflect a longer conversation history:

```
[
    {"role": "system", "content": "system message here"},
    {"role": "user", "content": "first user prompt here"},
    {"role": "assistant", "content": "the assistant's response"},
    {"role": "user", "content": "the new user prompt"},
]
```

And we can use this approach to engage in a longer interaction with history.


````python
# Let's make a conversation between GPT-4.1-mini and Claude-haiku-4.5
# We're using cheap versions of models so the costs will be minimal

gpt_model = "gpt-4.1-mini"
claude_model = "claude-haiku-4-5"

gpt_system = "You are a chatbot who is very argumentative; \
you disagree with anything in the conversation and you challenge everything, in a snarky way."

claude_system = "You are a very polite, courteous chatbot. You try to agree with \
everything the other person says, or find common ground. If the other person is argumentative, \
you try to calm them down and keep chatting."

gpt_messages = ["Hi there"]
claude_messages = ["Hi"]
````

````python
def call_gpt():
    messages = [{"role": "system", "content": gpt_system}]
    for gpt, claude in zip(gpt_messages, claude_messages):
        # zip(gpt_messages, claude_messages) pairs up items from both lists by position.
        # Here, it produces one pair: ("Hi there", "Hi"). if there's subsequent messages, then there will be ("Hi there", "Hi"), ("Something snarky", "Something polite") -> index 0 with index 0, index 1 with index 1
        # In this loop iteration, gpt gets "Hi there" and claude gets "Hi".
        # Then those values are appended to messages with the appropriate roles.
        messages.append({"role": "assistant", "content": gpt})
        # Rebuild the past conversation by pairing each GPT message
        # with the corresponding Claude message
        messages.append({"role": "user", "content": claude})
    response = openai.chat.completions.create(model=gpt_model, messages=messages)
    return response.choices[0].message.content
````

````python
# What does zip do.
messages_test = [{"role": "system", "content": gpt_system}]
for gpt, claude in zip(gpt_messages, claude_messages):
    print(messages_test)
    # what is this zip function doing?
    messages_test.append({"role": "assistant", "content": gpt})
    # Rebuild the past conversation by pairing each GPT message
    # with the corresponding Claude message
    messages_test.append({"role": "user", "content": claude})
````

````python
call_gpt()
````

````python
def call_claude():
    messages = [{"role": "system", "content": claude_system}]
    for gpt, claude_message in zip(gpt_messages, claude_messages):
        messages.append({"role": "user", "content": gpt})
        messages.append({"role": "assistant", "content": claude_message})
    messages.append({"role": "user", "content": gpt_messages[-1]})
    response = anthropic.chat.completions.create(model=claude_model, messages=messages)
    return response.choices[0].message.content
````

````python
call_claude()
````

````python
call_gpt()
````

````python
gpt_messages = ["Hi there"]
claude_messages = ["Hi"]

display(Markdown(f"### GPT:\n{gpt_messages[0]}\n"))
display(Markdown(f"### Claude:\n{claude_messages[0]}\n"))

for i in range(5):
    gpt_next = call_gpt()
    display(Markdown(f"### GPT:\n{gpt_next}\n"))
    gpt_messages.append(gpt_next)

    claude_next = call_claude()
    display(Markdown(f"### Claude:\n{claude_next}\n"))
    claude_messages.append(claude_next)
````

# More advanced exercises

Try creating a 3-way, perhaps bringing Gemini into the conversation! One student has completed this - see the implementation in the community-contributions folder.

The most reliable way to do this involves thinking a bit differently about your prompts: just 1 system prompt and 1 user prompt each time, and in the user prompt list the full conversation so far.

Something like:

```python
system_prompt = """
You are Alex, a chatbot who is very argumentative; you disagree with anything in the conversation and you challenge everything, in a snarky way.
You are in a conversation with Blake and Charlie.
"""

user_prompt = f"""
You are Alex, in conversation with Blake and Charlie.
The conversation so far is as follows:
{conversation}
Now with this, respond with what you would like to say next, as Alex.
"""
```

Try doing this yourself before you look at the solutions. It's easiest to use the OpenAI python client to access the Gemini model (see the 2nd Gemini example above).

Can also have the same model but have different prompts for different personas! so just 1 model can also use this trick for a 3-way conversation instead of user -> assistant -> user -> assistant
